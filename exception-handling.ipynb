```python
#try block
a = int(input('Enter first no.'))
b = int(input('Enter second no.'))

try:
    print(a/b)

#exception block
except Exception:
    print('divide by zero is not possible')

print('Bye')
```

```python

a = 10
b = 0

#try block
try:
    print('Resource in open')
    print(a/b)

#exception block
except Exception as e:
    print('divide by zero is not possible....',e)

finally:
    print('Resource closed') #this block will execute every time.
```
